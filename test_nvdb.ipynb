```python
import pandas as pd
import requests
```

```python
def get_vlseq(link:str,start:str):
    params = {'srid':'4326',
              'vegsystemreferanse':'EV14', #change vegsystemreferanse to the FV number or EV number you want
              'alle_versjoner':'true', 'inkluder': 'alle',
              'start': start}
    resp = requests.get(link, 
                    params=params)
    return resp.json()
```

```python
BASE_API_URL = 'https://nvdbapiles-v3.atlas.vegvesen.no'
signal_pts = BASE_API_URL + '/vegobjekter/96' #96 for traffic signs
respjson = get_vlseq(link=signal_pts,start=None)
objekter = respjson['objekter']
df_vsq = pd.DataFrame.from_records(objekter)
# egenskaper = pd.DataFrame.from_records(df_vsq['egenskaper'])
df_vsq
```

```python
df_org = pd.json_normalize(respjson['objekter']) #splits only 2 level lists in to separate columns

df_sk = pd.DataFrame(columns=['Skiltnummer'])
for idx, list in enumerate(df_org['egenskaper']):
    for e in list:
        if e['navn'] == 'Skiltnummer':
            # print('navn: ',e['navn'],' verdi: ',e['verdi'])
            df_sk.at[idx,'Skiltnummer'] = e['verdi']
df = df_org[['id','href','geometri.wkt']]
df = df.join(df_sk)
df


```
